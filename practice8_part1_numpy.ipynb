### Compare the time it takes to multiply elements of two lists pairwise using numpy and base python. Specifically make two lists with 1 million elements each with random entries from the interval (0,1).

```python
import time
import random
A = []
B = []
for i in range(1000000):
    A.append(random.random())
    B.append(random.random())
start = time.time()
Prod1 = np.array(A) * np.array(B)
Prod2 = []
end = time.time()
for i in range(1000000):
    Prod2.append(A[i]*B[i])
end2 = time.time()
print(f'With numpy: {end-start}\nBase python: {end2-start}')
```

### Use a mask to black out the entire image except the racoon's face. The face is at about the coordinates 660,330 and its radius is about 200 units.

```python
center_x, center_y = 660, 330
radius = 200

indices = np.indices(face.shape[:2])

dist_sq = (indices[0]-center_y)**2+(indices[1]-center_x)**2
mask = dist_sq > radius**2
mask
#mask is applied when its false
#apply the mask
face[mask] = 0
plt.imshow(face)
```

### Loading, plotting and analyzing simple data

The file stockholm_temperatures.dat contains the average daily temperature in Stockholm since 1800 until 2011. The first three columns are respectively year, month and day, while the following column is the temperature.



```python
data = np.loadtxt('stockholm_temperatures.dat')
data[0]
```

### Over 200 years, what is the mean temperature in Stockholm?

```python
np.mean(data[:,3]),np.std(data[:,3]),np.var(data[:,3])
# ':' takes all the rows
# , 3 takes the 4th column
```

### What is the highest average daily temperature recorded?

```python
np.max(data[:,3]), data[:,3].max()
```

### Use a mask to extract temperatures from 1987, only. Define the mask
mask = (data[X,Y] -condition-)

Where X defines the rows you want to consider (should be all), and Y defines the column you want to consider (only the year column), -condition- is the filter.


```python
mask = data[:,0] == 1987
data[mask, 3]
```

### Here is code to plot the average monthly temperature. Can you modify it to print the average yearly temperature?

```python
monthly_mean = [np.mean(data[data[:,1] == month, 3]) for month in range(1,13)]

plt.bar(range(1,13), monthly_mean)
plt.ylabel('Average Temperature')
plt.xlabel('Month')
```

```python
yearly_mean = [np.mean(data[data[:,0] == year, 3]) for year in range(1800,2012)]

plt.bar(range(1800,2012), yearly_mean)
plt.ylabel('Average Temperature')
plt.xlabel('Month')
```
