## Exercise 1 

Write a function to sum the squares of the values of a list, once with and once without list comprehension. Use the %%timeit function to test the speed of the two implementations on a randomly generated list of number of length 1000.

```python
%%timeit
import random

l = []
for i in range(1000):
    l.append(random.randint(1,100))

def summing(list):
    szum = 0
    for i in list:
        szum += i**2
    return szum

print(summing(l))
```
